```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')

feature_cols = df.select_dtypes(include=[np.number]).columns.drop('quality', errors='ignore')
df[feature_cols] = df.groupby('type')[feature_cols].transform(
    lambda x: imputer.fit_transform(x.values.reshape(-1, 1)).ravel()
    )
```

```python
df.info()
```

```python
plt.figure(figsize=(30,20))
corr = df.select_dtypes(include=[np.number]).corr()
sb.heatmap(corr, annot=True)
plt.show()
```

```python
from scipy import stats

numeric_df = df.select_dtypes(include=[np.number])
z_scores = np.abs(stats.zscore(numeric_df))
print(z_scores)
```

```python
outlier_counts = (z_scores > 3).sum(axis=1)
df_filtered = df[outlier_counts <= 2]
df_filtered.shape
```
